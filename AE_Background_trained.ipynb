```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('synthetic_dataset.csv')

# Separate features and labels
X = df.drop(columns=['Class'])  # Features
y = df['Class']  # Labels

# List of feature combinations for plotting
feature_combinations = [('Feature 1', 'Feature 2'), ('Feature 1', 'Feature 3'),
                        ('Feature 1', 'Feature 4'), ('Feature 2', 'Feature 3'),
                        ('Feature 2', 'Feature 4'), ('Feature 3', 'Feature 4')]

# Plot each feature combination
for i, (feature1, feature2) in enumerate(feature_combinations, start=1):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[y == 0][feature1], X[y == 0][feature2], color='blue', label='Class 1', alpha=0.5)
    plt.scatter(X[y == 1][feature1], X[y == 1][feature2], color='red', label='Class 2', alpha=0.5)
    plt.xlabel(feature1)
    plt.ylabel(feature2)
    plt.title(f'{feature1} vs {feature2}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the normal dataset
normal_df = pd.read_csv('normal_dataset.csv')

# Split the normal dataset into train and test sets (80% train, 20% test)
normal_train_dataset, normal_test_dataset = train_test_split(normal_df, test_size=0.2, random_state=42)

# Save the normal train and test datasets to CSV files
normal_train_dataset.to_csv('normal_train_dataset.csv', index=False)
normal_test_dataset.to_csv('normal_test_dataset.csv', index=False)

print("Normal dataset split into train and test subsets and saved successfully.")

# Load the anomaly dataset
anomaly_df = pd.read_csv('anomaly_dataset.csv')

# Split the anomaly dataset into train and test sets (80% train, 20% test)
anomaly_train_dataset, anomaly_test_dataset = train_test_split(anomaly_df, test_size=0.2, random_state=42)

# Save the anomaly train and test datasets to CSV files
anomaly_train_dataset.to_csv('anomaly_train_dataset.csv', index=False)
anomaly_test_dataset.to_csv('anomaly_test_dataset.csv', index=False)

print("Anomaly dataset split into train and test subsets and saved successfully.")
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the datasets
normal_train_data = pd.read_csv('normal_train_data.csv')
normal_test_data = pd.read_csv('normal_test_data.csv')
anomaly_train_data = pd.read_csv('anomaly_train_data.csv')
anomaly_test_data = pd.read_csv('anomaly_test_data.csv')

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Fit scaler on training data and transform both training and testing data
normal_train_data_scaled = scaler.fit_transform(normal_train_data)
normal_test_data_scaled = scaler.transform(normal_test_data)
anomaly_train_data_scaled = scaler.transform(anomaly_train_data)
anomaly_test_data_scaled = scaler.transform(anomaly_test_data)

# Convert scaled arrays back to DataFrames
normal_train_data_scaled = pd.DataFrame(normal_train_data_scaled, columns=normal_train_data.columns)
normal_test_data_scaled = pd.DataFrame(normal_test_data_scaled, columns=normal_test_data.columns)
anomaly_train_data_scaled = pd.DataFrame(anomaly_train_data_scaled, columns=anomaly_train_data.columns)
anomaly_test_data_scaled = pd.DataFrame(anomaly_test_data_scaled, columns=anomaly_test_data.columns)

# Save the scaled data to new CSV files
normal_train_data_scaled.to_csv('normal_train_data_scaled.csv', index=False)
normal_test_data_scaled.to_csv('normal_test_data_scaled.csv', index=False)
anomaly_train_data_scaled.to_csv('anomaly_train_data_scaled.csv', index=False)
anomaly_test_data_scaled.to_csv('anomaly_test_data_scaled.csv', index=False)

print("Data normalization using MinMaxScaler completed and saved successfully.")

```

```python
import tensorflow as tf
from tensorflow.keras.models import Model

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(4,activation='relu'))
model.add(tf.keras.layers.Dense(2,activation='relu'))
model.add(tf.keras.layers.Dense(4,activation='sigmoid'))
```

```python
class AutoEncoder(Model):
    def __init__(self):
        super(AutoEncoder, self).__init__()
        self.encoder = tf.keras.Sequential([
          tf.keras.layers.Dense(4,activation="relu"),
          tf.keras.layers.Dense(2,activation="relu")])

        self.decoder = tf.keras.Sequential([
          tf.keras.layers.Dense(4,activation="sigmoid")])

    def call(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
```

```python
model = AutoEncoder()
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                                  patience=2,
                                                  mode='min')

model.compile(optimizer='adam', loss='mae')
```

```python
history = model.fit(normal_train_data_scaled,normal_train_data_scaled,
                    epochs=50,
                    batch_size=128,
                    validation_data= (normal_test_data_scaled,normal_test_data_scaled),
                    shuffle=True,
                    callbacks=[early_stopping])
```
